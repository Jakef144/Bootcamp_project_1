```python
# Define the API endpoint and parameters
url = f"https://api.geoapify.com/v2/places?categories=healthcare&filter=rect:-79.9449,42.2696,-78.8995,41.8419&limit=100&keywords=addiction,rehabilitation,substance%20abuse,recovery,treatment&apiKey={geo_api}"

headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"

# Make the request
resp = requests.get(url, headers=headers)

if resp.status_code == 200:
    data = resp.json()
    places = data.get("features", [])
    
    # Process the results
    if places:
        print(f"List of Addiction Treatment Centers Found in Erie County (Total: {len(places)}):")
        for i, place in enumerate(places, start=1):
            properties = place.get("properties", {})
            name = properties.get("name", "No name available")
            address = properties.get("address_line1", "No address available")
            city = properties.get("city", "No city available")
            category = properties.get("categories", [])
            
            print(f"{i}. Name: {name}")
            print(f"   Address: {address}")
            print(f"   City: {city}")
            print(f"   Categories: {', '.join(category)}\n")
    else:
        print("No addiction treatment centers found.")
else:
    print(f"Error {resp.status_code}: {resp.text}")

```

```python
import requests

# Overpass API query to find substance abuse-related healthcare facilities in Erie County
query = """
[out:json];
area[name="Erie County"][admin_level=6];
node["amenity"="clinic"](area);
out body;
"""

url = "http://overpass-api.de/api/interpreter"
response = requests.get(url, params={'data': query})

if response.status_code == 200:
    data = response.json()
    elements = data.get("elements", [])
    
    if elements:
        print(f"Found {len(elements)} facilities:")
        for i, element in enumerate(elements, start=1):
            tags = element.get("tags", {})
            name = tags.get("name", "No name available")
            address = tags.get("addr:street", "No address available")
            city = tags.get("addr:city", "No city available")
            facility_type = tags.get("healthcare", "No type available")
            
            print(f"{i}. Name: {name}")
            print(f"   Address: {address}")
            print(f"   City: {city}")
            print(f"   Facility Type: {facility_type}\n")
    else:
        print("No relevant facilities found.")
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
# Overpass API query to find clinics or healthcare facilities in Erie County
query = """
[out:json];
area[name="Erie County"][admin_level=6];
node["amenity"="clinic"](area);
out body;
"""

url = "http://overpass-api.de/api/interpreter"
response = requests.get(url, params={'data': query})

if response.status_code == 200:
    data = response.json()
    elements = data.get("elements", [])
    
    # Filter to show only facilities in Erie County
    erie_county_facilities = [
        element for element in elements 
        if 'Erie' in (element.get("tags", {}).get("addr:city", "") or element.get("tags", {}).get("name", ""))
    ]
    
    if erie_county_facilities:
        print(f"Found {len(erie_county_facilities)} facilities in Erie County:")
        for i, element in enumerate(erie_county_facilities, start=1):
            tags = element.get("tags", {})
            name = tags.get("name", "No name available")
            address = tags.get("addr:street", "No address available")
            city = tags.get("addr:city", "No city available")
            facility_type = tags.get("healthcare", "No type available")
            
            print(f"{i}. Name: {name}")
            print(f"   Address: {address}")
            print(f"   City: {city}")
            print(f"   Facility Type: {facility_type}\n")
    else:
        print("No facilities found in Erie County.")
else:
    print(f"Error {response.status_code}: {response.text}")
```
