```python
import pandas as pd
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import folium
```

```python
# Load the data with a specified encoding
file_path = 'CSV_files/National Directory SU 2024_Final.xlsx'
data = pd.read_excel(file_path)

# Display the column headers
print(data.columns)
# Display the first few rows of data to check content
print(data.head())
```

```python
# Filtering the data to keep only facilities in New York (state code 'NY')
ny_facilities = data[data['state'] == 'NY']

# Displaying
ny_facilities.head()

```

```python
# Renaming the 'name1' column to 'Facility Name'
ny_facilities = ny_facilities.rename(columns={'name1': 'Facility Name'})
ny_facilities = ny_facilities.rename(columns={'name2': 'Type of Clinic'})
ny_facilities = ny_facilities.drop(columns=['street2', 'phone', 'intake1', 'intake2', 'intake1a', 'intake2a'])

# Displaying 
ny_facilities.head()

```

```python
# Match the earlier state
ny_facilities = data[data['state'] == 'NY']
ny_facilities = ny_facilities.rename(columns={'name1': 'Facility Name', 'name2': 'Type of Clinic'})
ny_facilities = ny_facilities.drop(columns=['street2', 'phone', 'intake1', 'intake2', 'intake1a', 'intake2a'])

# Filtering 'service_code_info' to include only rows with specified codes
valid_codes = {'SA', 'HH', 'SUMH', 'OP', 'RES', 'RL', 'RS', 'OTP', 'BIA', 'CRV', 
               'CMMI', 'CBT', 'RELP', 'SACA', 'TRC', 'TWFA', 'TELE', 
               'STAG', 'SHG', 'ICO', 'GCO', 'FCO', 'AUDO', 'SPS'}

ny_facilities_filtered = ny_facilities[
    ny_facilities['service_code_info'].apply(
        lambda x: any(code in x for code in valid_codes) if isinstance(x, str) else False
    )
]

# Displaying the filtered DataFrame
ny_facilities_filtered.head()

```

```python
# Filtering facilities based on the provided list of ZIP codes
valid_zips = [
    14221, 14075, 14215, 14150, 14224, 14225, 14086, 14127, 14226, 14228,
    14207, 14043, 14220, 14211, 14072, 14213, 14216, 14217, 14206, 14227,
    14214, 14223, 14051, 14218, 14052, 14222, 14210, 14201, 14219, 14004,
    14208, 14212, 14031, 14032, 14001, 14059, 14006, 14057, 14209, 14141,
    14204, 14085, 14068, 14070, 14047, 14080, 14202, 14203, 14111, 14025,
    14081, 14034, 14033, 14170, 14139, 14055, 14091, 14030, 14102, 14037,
    14026, 14069, 14134, 14061, 14110, 14112, 14140, 14151, 14169, 14205,
    14010, 14027, 14035, 14038, 14231, 14233, 14240, 14241, 14260, 14261,
    14263, 14264, 14265, 14267, 14269, 14270, 14272, 14273, 14276, 14280
]

# Filtering the facilities based on the valid ZIP codes
filtered_facilities_by_zip = ny_facilities_filtered[
    ny_facilities_filtered['zip'].astype(str).astype(int).isin(valid_zips)
]

# Displaying the count of facilities that match the criteria
filtered_facilities_by_zip
```

```python
filtered_count = len(filtered_facilities_by_zip)
filtered_count
```

```python
from geopy.geocoders import Nominatim
import time

# Initialize the geolocator
geolocator = Nominatim(user_agent="facility_mapper")

# Function to get coordinates for each address
def get_coordinates(address):
    try:
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        return None, None
    except:
        return None, None

# Create a full address column for geocoding
filtered_facilities_by_zip.loc[:, 'address'] = filtered_facilities_by_zip['street1'] + ', ' + filtered_facilities_by_zip['city'] + ', ' + filtered_facilities_by_zip['state']

# Apply the geocoding function to each address
filtered_facilities_by_zip.loc[:, ['latitude', 'longitude']] = pd.DataFrame(
    filtered_facilities_by_zip['address'].apply(get_coordinates).tolist(), 
    index=filtered_facilities_by_zip.index
)

# Add a small delay between requests to respect the API's rate limits
time.sleep(1)
```

```python
# Displaying the DataFrame with the updated coordinates
filtered_facilities_by_zip[['Facility Name', 'address', 'latitude', 'longitude']]
for address in filtered_facilities_by_zip['address']:
    print(f"Geocoding address: {address}")
    lat, lon = get_coordinates(address)
    print(f"Result: Latitude: {lat}, Longitude: {lon}")
```

```python
# Iterate over each address and update the DataFrame with geocoded values
for index, address in filtered_facilities_by_zip['address'].items():
    lat, lon = get_coordinates(address)
    filtered_facilities_by_zip.loc[index, 'latitude'] = lat
    filtered_facilities_by_zip.loc[index, 'longitude'] = lon

# Display the updated DataFrame
filtered_facilities_by_zip[['Facility Name', 'address', 'latitude', 'longitude']]
```

```python
# Creating a Folium map for the facilities with updated coordinates
map_updated_facilities = folium.Map(location=[42.8864, -78.8784], zoom_start=10)

# Adding smaller circle markers with precise coordinates
for index, row in filtered_facilities_by_zip.iterrows():
    if not pd.isnull(row['latitude']) and not pd.isnull(row['longitude']):
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=5,  # Smaller radius for the marker
            color='blue',
            fill=True,
            fill_color='blue',
            fill_opacity=0.7,
            popup=f"{row['Facility Name']}<br>{row['Type of Clinic']}<br>{row['street1']}, {row['city']} {row['zip']}",
            tooltip=row['Facility Name']
        ).add_to(map_updated_facilities)

# Displaying the map
map_updated_facilities
```
