```python
import pandas as pd
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import folium
```

```python
# Load the data with a specified encoding
file_path = 'CSV_files/National Directory SU 2024_Final.xlsx'
data = pd.read_excel(file_path)

# Display the column headers
print(data.columns)
# Display the first few rows of data to check content
print(data.head())
```

```python
# Filtering the data to keep only facilities in New York (state code 'NY')
ny_facilities = data[data['state'] == 'NY']

# Displaying
ny_facilities.head()
```

```python
# Renaming the 'name1' column to 'Facility Name'
ny_facilities = ny_facilities.rename(columns={'name1': 'Facility Name'})
ny_facilities = ny_facilities.rename(columns={'name2': 'Type of Clinic'})
ny_facilities = ny_facilities.drop(columns=['street2', 'phone', 'intake1', 'intake2', 'intake1a', 'intake2a'])

# Displaying 
ny_facilities.head()
```

```python
# Match the earlier state
ny_facilities = data[data['state'] == 'NY']
ny_facilities = ny_facilities.rename(columns={'name1': 'Facility Name', 'name2': 'Type of Clinic'})
ny_facilities = ny_facilities.drop(columns=['street2', 'phone', 'intake1', 'intake2', 'intake1a', 'intake2a'])

# Filtering 'service_code_info' to include only rows with specified codes
valid_codes = {'SA', 'HH', 'SUMH', 'OP', 'RES', 'RL', 'RS', 'OTP', 'BIA', 'CRV', 
               'CMMI', 'CBT', 'RELP', 'SACA', 'TRC', 'TWFA', 'TELE', 
               'STAG', 'SHG', 'ICO', 'GCO', 'FCO', 'AUDO', 'SPS'}

ny_facilities_filtered = ny_facilities[
    ny_facilities['service_code_info'].apply(
        lambda x: any(code in x for code in valid_codes) if isinstance(x, str) else False
    )
]

# Displaying the filtered DataFrame
ny_facilities_filtered.head()
```

```python
# List of Westchester ZIP codes
westchester_zips = [
    10701, 10801, 10573, 10705, 10583, 10550, 10704, 10562, 10598, 10710, 10558, 10566, 10557,
    10708, 10552, 10703, 10567, 10605, 10543, 10591, 10603, 10805, 10549, 10580, 10804, 10538,
    10606, 10528, 10514, 10530, 10803, 10604, 10553, 10570, 10536, 10601, 10520, 10707, 10522,
    10523, 10595, 10510, 10709, 10547, 10706, 10589, 10504, 10533, 10590, 10502, 10607, 10507,
    10560, 10506, 10576, 10532, 10577, 10594, 10548, 10588, 10537, 10511, 10526, 10527, 10597,
    10518, 10501, 10596, 10578, 10546, 10505, 10535, 10571, 10572, 10540, 10545, 10551, 10587,
    10602, 10610, 10702, 10802, 10503, 10517, 10519, 10521
]

# Filtering for facilities in Westchester based on the provided ZIP codes
westchester_facilities = ny_facilities_filtered[
    ny_facilities_filtered['zip'].astype(str).astype(int).isin(westchester_zips)
]

# Displaying the filtered facilities
westchester_facilities.head()
```

```python
from geopy.geocoders import Nominatim
import time

# Initialize the geolocator
geolocator = Nominatim(user_agent="facility_mapper")

# Function to get coordinates for each address
def get_coordinates(address):
    try:
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        return None, None
    except:
        return None, None

# Create a new column for addresses without triggering a SettingWithCopyWarning
westchester_facilities = westchester_facilities.copy()  # Create a copy to avoid modifying a view
westchester_facilities['address'] = westchester_facilities['street1'] + ', ' + westchester_facilities['city'] + ', ' + westchester_facilities['state']

# Apply the geocoding function with a loop and update columns safely
latitude_list = []
longitude_list = []
for address in westchester_facilities['address']:
    lat, lon = get_coordinates(address)
    latitude_list.append(lat)
    longitude_list.append(lon)
    time.sleep(1)  # Respect rate limits

# Assign the latitude and longitude lists to new columns
westchester_facilities['latitude'] = latitude_list
westchester_facilities['longitude'] = longitude_list

# Display the updated DataFrame
westchester_facilities[['Facility Name', 'address', 'latitude', 'longitude']]
```

```python
# Displaying the DataFrame with the updated coordinates
westchester_facilities[['Facility Name', 'address', 'latitude', 'longitude']]
for address in westchester_facilities['address']:
    print(f"Geocoding address: {address}")
    lat, lon = get_coordinates(address)
    print(f"Result: Latitude: {lat}, Longitude: {lon}")
```

```python
# Creating a Folium map for the Westchester facilities with smaller indicators

# Generating a base map centered around Westchester County, NY
map_westchester_facilities = folium.Map(location=[41.1220, -73.7949], zoom_start=10)

# Adding smaller circle markers for facilities with valid latitude and longitude values
for index, row in westchester_facilities.iterrows():
    if not pd.isnull(row['latitude']) and not pd.isnull(row['longitude']):
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=5,  # Smaller radius for the indicator
            color='blue',
            fill=True,
            fill_color='blue',
            fill_opacity=0.7,
            popup=f"{row['Facility Name']}<br>{row['Type of Clinic']}<br>{row['street1']}, {row['city']} {row['zip']}",
            tooltip=row['Facility Name']
        ).add_to(map_westchester_facilities)

# Displaying the map
map_westchester_facilities

```
