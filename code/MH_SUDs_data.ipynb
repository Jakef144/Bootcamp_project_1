```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import io

headers = {
	"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

metadata_url = "https://ourworldindata.org/grapher/death-rates-from-mental-and-substance-disorders-by-age-who.metadata.json?v=1&csvType=full&useColumnShortNames=true"
metadata = requests.get(metadata_url, headers=headers).json()

csv_url = "https://ourworldindata.org/grapher/death-rates-from-mental-and-substance-disorders-by-age-who.csv?v=1&csvType=full&useColumnShortNames=true"
response = requests.get(csv_url, headers=headers)
response.raise_for_status()  # Check if the request was successful
df = pd.read_csv(io.StringIO(response.text))
```

```python
df.to_csv('death_rates_mental_substance_disorders.csv', index=False)
```

```python
# Filter the dataframe for the "United States" entity
us_df = df[df['Entity'] == 'United States']

# Define the years you want to filter
filtered_years = [2000, 2005, 2010, 2015, 2020, 2021]

# Filter the dataframe for the "United States" entity and the specified years
us_filtered_df = us_df[us_df['Year'].isin(filtered_years)]

# Plotting the bar chart for the "United States" entity

us_filtered_df.plot(x='Year', y=[
    'Age 0-4',
    'Age 5-14',
    'Age 15-49',
    'Age 50-69',
    'Age 70+',
    'All Ages'
], kind='bar', figsize=(14, 8))

plt.title('Death Rates from Mental and Substance Use Disorders by Age Group (United States)')
plt.xlabel('Year')
plt.ylabel('Death Rate per 100k')
plt.legend(title='Age Groups')
plt.show()
```
